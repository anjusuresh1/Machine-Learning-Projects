```python
#Importing the dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('uber-raw-data-sep14.csv')
```

```python
data.head()
```

```python
data['Date/Time']=data['Date/Time'].map(pd.to_datetime)
```

```python
data.head()
```

```python
data['Day']=data['Date/Time'].apply(lambda x: x.day)
data['WeekDay']=data['Date/Time'].apply(lambda x: x.weekday())
data['Hour']=data['Date/Time'].apply(lambda x: x.hour)
data.head()
```

```python
newData = data.drop('Week Day', axis=1)
newData.head()
```

```python
sns.displot(data['Day'])
```

Distribution Plot shows that the Uber rides increases during week days and rides request reduces during weekend.

```python
sns.displot(data['Hour'])
```

The uber rides rises from 5 am onwards and reaches its peak around 6pm, then slowly decreases from midnight onwards.

```python
sns.displot(data['WeekDay'])
```

0 - sunday
Among the weekends, Sunday is having more rides compared to Saturday./
Monday and Friday is topping among the weekdays when it comes to uber rides.


```python
df = data.groupby(['WeekDay','Hour']).apply(lambda x: len(x))
df =df.unstack()
sns.heatmap(df, annot= False)
```

Conclusions:
Monday is the most profitable day for Uber
People use less uber during Saturday
The rise in uber ride starts from 5am onwards
The most peak time of uber rides is 6pm
