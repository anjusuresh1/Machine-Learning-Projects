#IMDB Movie Rating WebScraping using Python

###Importing the dependencies

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import json
```

###Fetching Url content

```python
url = 'http://www.imdb.com/chart/top/'
response = requests.get(url)

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Accept-Language": "en-US,en;q=0.9",
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
print(response.text[:1000])
```

###Fetching the required elements from the url content

```python
scripts = soup.find_all("script", type="application/ld+json")

movies = []

for script in scripts:
    data = json.loads(script.string)
    if data.get("@type") == "ItemList":
        movies = data["itemListElement"]
        break

print(len(movies))
```

```python
movie_data = []
for movie in movies:
  movie_data.append({
      'title': movie['item']['name'],
      'rating': movie['item']['aggregateRating']['ratingValue'],
      'genre': movie['item']['genre']
  })
print(movie_data[0])
```

```python
for movie in movie_data:
  print(f"{movie['title']} {movie['rating']} - {movie['genre']}")
```

###Store the extracted info in cv

```python
df = pd.DataFrame(movie_data)
df.to_csv('IMDB_Top_Movies_Info.csv', index= False)
print('The saving of top IMDB movie using web scraping is completed')
```
